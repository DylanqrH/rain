#  **Predicting Rainfall Tomorrow in Darwin, Australia**


```r
#Loading appropriate packages into the notebook
library(tidyverse)
library(repr)
library(tidymodels)
options(repr.matrix.max.rows = 6)
```

### **Introduction**
   Rainfall in Australia is highly varied with low average annual rainfall over most of the continent, and intense seasonal rainfall in the tropical areas.

   “Rain in Australia” dataset contains daily weather observations ranging from 2007 to 2017, collected from several locations across Australia. With this dataset, we are trying to answer the following question:

**Will it rain the next day in Darwin?**

We start off by reading the dataset into the notebook:

```r
#Reading the dataset into the notebook
weather <- read_csv("https://raw.githubusercontent.com/DylanqrH/rain/main/weatherAUS.csv", guess_max = 10000)
weather
```

To answer this question, we first filtered the location for Darwin, and selected 8 of the most significant variables from this dataset. We cross-checked the dataset with the government source to make sure that the data had not been falsified and that we could use it to train our model. With the selected variables, we can predict the next-day rain by training a classification model on a target variable.

```r
weather_Darwin <- weather %>%
                  filter(Location == "Darwin") %>%
                  select(Sunshine, WindDir9am, Humidity3pm, Pressure3pm, Rainfall, MinTemp, MaxTemp, RainTomorrow) %>%
                  mutate(RainTomorrow = as.factor(RainTomorrow)) %>%
                  mutate(WindDir9am = as.factor(WindDir9am)) %>%
                  drop_na()

weather_Darwin
```

Next, we split our data into a training and testing set using `initial_split`. We will carry out our visualizations on the training set and keep the testing set to test our accuracy at the end.

```r
set.seed(999)

weather_split <- initial_split(weather_Darwin, prop = 0.75, strata = RainTomorrow)  
weather_train <- training(weather_split)   
weather_test <- testing(weather_split)
weather_train
weather_test
```

### **Visualizations**

To analyse our data, we will be utilizing the $K$-nearest neighbours classification model to determine whether or not it will rain the next day in Darwin. We researched about factors that have the strongest effect on next day rainfall - helping us extract the following variables for our visualization: `Sunshine`, `Humidity3pm`, `Pressure3pm`, `Rainfall`, `MinTemp`.

We illustrated multiple graphs to see the relationships and effects of variables on `RainTomorrow`.


```r
options(repr.plot.width = 11, repr.plot.height = 8)

weather_plot <- weather_train %>%
    ggplot(aes(x = Pressure3pm, y = Humidity3pm, color = RainTomorrow)) + 
        geom_point(alpha = 0.5) +
        labs(x = "Pressure at 3pm (hPa)", y = "Humidity at 3pm (%)", color = "Will It Rain Tomorrow?", 
             title = "The Effect of Pressure and Humidity on Next Day’s Weather") +
        theme(text = element_text(size = 18))
        
weather_plot
```

With our first scatterplot, we found that with higher humidity and lower pressure, there is a higher chance that it will rain tomorrow - vice versa. We decided to use humidity and pressure at 3pm instead of 9am since it is closer to the next day's weather, giving a better prediction.

```r
options(repr.plot.width = 11, repr.plot.height = 8)

weather_humid_temp_plot <- weather_train %>%
    ggplot(aes(x = MinTemp, y = Humidity3pm, color = RainTomorrow)) + 
        geom_point(alpha = 0.5) +
        labs(x = "Minimum Temperature Today (ºC)", y = "Humidity at 3pm (%)", color = "Will it Rain Tomorrow?") +
        ggtitle("How Does the Minimum Temperature Effect the Humidity?") + 
        theme(text = element_text(size = 18))

weather_humid_temp_plot
```

When observing the effects of minimum temperature of the day to humidity at 3pm, it can be noted that as humidity and minimum temperature decrease, the likelihood that it will rain also decreases. We concluded that temperature has a significant influence on the amount of humidity. Therefore, we are using both `MaxTemp` and `MinTemp` as predictors.

```r
options(repr.plot.width = 11, repr.plot.height = 8)

weather_sunshine <- weather_train %>%
    ggplot(aes(x = Sunshine, fill = RainTomorrow)) + 
        geom_histogram() +
        facet_grid(rows = vars(RainTomorrow))+
        labs(x = "Hours of Bright Sunshine per Day", y = "Number of Occurences", fill = "Will It Rain Tomorrow?",
             title = "Hours of Bright Sunshine Against the Likelihood of it Raining the Next Day") + 
        theme(text = element_text(size = 18))
weather_sunshine
```

Graphing the hours of sunshine against the days with rain the next day, we get two graphs for it raining versus not raining. We see that with increasing hours of sunshine, the chance of it raining the next day decreases. Hence, we selected the column `Rainfall`.
